```python
import pandas as pd
import numpy as np
```

```python
alignments_melo_csv = pd.read_csv(
    '/home/lab205/Desktop/LLUIS/Analysis/03-miRNA-hairpin-alignment/alignments_cmelo.csv', sep = "\t")
```

```python
alignments_melo_csv
```

```python
aligned_miRNA_mature = pd.DataFrame(alignments_melo_csv.iloc[:,[0,2,3]][121:])
aligned_miRNA_mature.columns = ['id miRNA', 'ID precursor', 'first base aligned']
aligned_miRNA_mature
```

```python
lengths_precursors = pd.DataFrame(alignments_melo_csv.iloc[:,[1,2]][:120])
```

```python
keys = []
values = []

for i in range(120):
    number = str(lengths_precursors.loc[[i,]])
    length_miRNA = number.split(':')[-1:][0]
    id_miRNA = number.split(':')[3].split()[0]
    keys.append(id_miRNA)
    values.append(length_miRNA)
```

```python
number = str(lengths_precursors.loc[[1,]])
number.split(':')[-1:][0]
```

```python
number.split(':')[3].split()[0]
```

```python
dictionary_precursors=dict(zip(keys, values))
dictionary_precursors
```

```python
aligned_miRNA_mature['length prec'] = [0]*len(aligned_miRNA_mature)
```

```python
for key in dictionary_precursors:
    for key2 in aligned_miRNA_mature['ID precursor']:
        if key == key2:
            indeces = [i for i, x in enumerate(list(aligned_miRNA_mature['ID precursor'])) if x == key]
            for index in indeces:
                aligned_miRNA_mature['length prec'].iloc[index,] = dictionary_precursors[key]
```

```python
aligned_miRNA_mature
############ GREAT JOB #######################
```

```python
aligned_miRNA_mature['%length'] = [None]*len(aligned_miRNA_mature)
aligned_miRNA_mature['5P or 3P'] = [None]*len(aligned_miRNA_mature)
counter_5P = 0
counter_3P = 0

for element in range(len(aligned_miRNA_mature)):
    perc = int(list(aligned_miRNA_mature['first base aligned'])[element]) / int(
        list(aligned_miRNA_mature['length prec'])[element])
    aligned_miRNA_mature['%length'].iloc[element,] = perc
    if perc <= 0.5:
        aligned_miRNA_mature['5P or 3P'].iloc[element,] = '5P'
        counter_5P += 1
    else:
        aligned_miRNA_mature['5P or 3P'].iloc[element,] = '3P'
        counter_3P += 1
```

```python
print(counter_5P, counter_3P)
```
